## 1. Описание данных
- `airports_nearest` — расстояние до ближайшего аэропорта в метрах (м)
- `balcony` — число балконов
- `ceiling_height` — высота потолков (м)

## 3. Предобработка данных

1. Переименование столбца:

```python
data.rename(columns = {'cityCenters_nearest': 'city_center_nearest'}, inplace=True)
data['city_center_nearest'] = round(data['city_center_nearest']/1000, 1).astype('int', errors='ignore')
```

2. Преобразование типов столбцов

Переведем столбец first_day_exposition к типу datatime без указания времени - его нет. 

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format = '%Y-%m-%d')
```

Заменим пропуски в balcony на 0 и сразу поменяем тип данных на int

```python
data['balcony'].value_counts()
```

```python
data['balcony'] = data['balcony'].fillna(0).astype('int')
data['balcony'].value_counts() #проверка
```

Посмотрим на значения столбца `days_exposition` и `first_day_exposition`.

```python
data[(pd.isna(data['days_exposition'])==True)]['first_day_exposition']
```

```python
data.query('first_day_exposition == first_day_exposition.max()')
```

Отсутсвие данных в столбце `days_exposition` говорит о том, что публикация еще не снята с объявления.  Не будем трогать такие значения: самая поздняя дата публикации объявления 2019-05-03 и там данные в изучаемом столбце равны NaN. Нулевых значений в столбце не наблюдалось. 

Мы можем заполнять данные только когда данные пропущены совершенно случайно (MCAR) - тогда заполнение не повлияет на характер связи между заполняемой переменной и остальными.

```python
# была попытка привести все к int через .values.astype(np.int64), но неудачно из-за nan. 
# Поэтому заменила все nan на ноль (других нулей в данных не было)
data['floors_total'] = data['floors_total'].fillna(0).astype('int')   
data['city_center_nearest'] = data['city_center_nearest'].fillna(0).astype('int')   
data['last_price'] = data['last_price'].astype('int')
data['total_area'] = data['total_area'].astype('float32')
data['ceiling_height'] = data['ceiling_height'].astype('float32')
data['living_area'] = data['living_area'].astype('float32')


data['parks_around3000'] = data['parks_around3000'].astype('float32')      
data['parks_nearest'] = data['parks_nearest'].astype('float32')
data['ponds_around3000'] = data['ponds_around3000'].astype('float32') 
data['ponds_nearest'] = data['ponds_nearest'].astype('float32')  
data['days_exposition'] = data['days_exposition'].astype('float32')
data['kitchen_area'] = data['kitchen_area'].astype('float32')
data['airports_nearest'] = data['airports_nearest'].astype('float32')

data.info() #проверка 
```

```python
pd.isna(data['kitchen_area']).sum()
```

Посмотрим пропуски в kitchen_area (2278): их можно было бы заменить также медианой произведения отношения площади кухни к общей площади или просто удалить, но также пропуски могут означать, что квартира-студия или со свободной планировкой (столбец atudio и open_plan). Проверим гипотезу.

```python
len(data.query('open_plan == True').query('kitchen_area.isna() == True', engine='python'))
```

```python
len(data.query('studio == True').query('kitchen_area.isna() == True', engine='python'))
```

Частично верно. Площадь кухни в современных квартирах с двумя и более комнатами, не может быть менее 8 кв. м, а в «однушках» – менее 5 кв. м. Дефолтным значением для студий и квартир со свободной планировкой оставим 0 (тк количество комнат равно нулю).

```python
data.loc[data['studio'] == True, 'kitchen_area'] = 0
data.loc[data['open_plan'] == True, 'kitchen_area'] = 0
data.loc[(data['rooms'] == 1)&(data['kitchen_area'].isna() == True), 'kitchen_area'] = 5
data.loc[(data['rooms'] != 1)&(data['kitchen_area'].isna() == True), 'kitchen_area'] = 8
```

```python
pd.isna(data['kitchen_area']).sum() #проверка
```

Теперь рассмотрим столбец `is_apartment`. Возможно, учет данного параметра ведется не с самого начала. Проверим гипотезу:

```python
data.sort_values(by='first_day_exposition').head(20)
```

```python
data['is_apartment'].value_counts()
```

Предположение неверно (есть и False, и Nan). Тогда исходим из здравого смысла - менее рискованно покупать и продавать не-аппартаменты, заменим все значения на  False. 

```python
data['is_apartment'] = data['is_apartment'].fillna(False).astype('bool')
```

Пропуски в living_area заполним медианой площади в зависимости от количесвта комнат:

```python
data['living_area'] = data['living_area'].fillna(data.groupby('rooms')['living_area'].transform('median'))
```

Удалим неизвестные данные в locality_name (менее 1%):

Поэтому пока оставим эти значения. То же касается и пропусков в ceiling_height - оставим эти значения.

```python
data = data.dropna(subset=['locality_name']).reset_index(drop=True)
pd.isna(data).sum() #проверка
```

```python
sorted(data['locality_name'].unique())
```

Что касается значений: происходит задвоение значений из-за е/ё и городского поселки и поселка городского типа.

```python
def replacer(df):
    for i in range(len(df)):
        df[i] = df[i].replace('ё', 'е')
        df[i] = df[i].replace('городской поселок', 'поселок городского типа')
    
replacer(data['locality_name'])
sorted(data['locality_name'].unique())
```

**Вывод**

Были изменены типы данных, заменены нулевые значения. Также изменила тип данных у площадей с float64 на float32 для экономии памяти

Нетронутыми остались значения в столбце `ceiling_hight`, `parks_nearest`, `ponds_nearest`, `parks_around3000`, `ponds_around3000`, `airports_nearest` и `city_center_nearest`.

## 4. Расчёты и добавление результатов в таблицу

Получим общую информацию о данных:
- цену квадратного метра;
- день недели, месяц и год публикации объявления;
- этаж квартиры; варианты — первый, последний, другой;
- соотношение жилой и общей площади, а также отношение площади кухни к общей.

```python
# Добавление цены квадратного метра:
data['price_per_m_scrt'] = round(data['last_price'] / data['total_area'], 2)

# Добавление дня недели, месяца и года публикации объявления:

data['weekday_publication'] = data['first_day_exposition'].dt.weekday
data['month_publication'] = data['first_day_exposition'].dt.month
data['year_publication'] = data['first_day_exposition'].dt.year

# Добавление этажа квартиры: варианты — первый, последний, другой:
# Создадим функцию: 
def flat_floor(row):
    try:
        if row['floor'] == 1:
            return 'первый'
        if row['floor'] == row['floors_total']:
            return 'последний'
        else: return 'другой'
    except: return "can't define"

data['flat_floor'] = data.apply(flat_floor, axis=1)

# Добавление соотношения жилой и общей площади:
data['ratio_liv_to_total_area'] = round(data['living_area']/data['total_area'], 3)

# а также отношение площади кухни к общей:
data['ratio_kit_to_total_area'] = round(data['kitchen_area']/data['total_area'], 3) 

data.head(20) #проверка
```
